```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

#read data
df = pd.read_csv('Education_attainment.csv')
df.head()
```

```python
#Determine the percentages for women vs. men having earned a Bachelor's Degree in
#1980.
data1980=df[(df['Year']==1980) &
                (df['Sex'] != 'A')&
               (df['Min degree'] == "bachelor's")]
[['Year','Sex','Min degree','Total','White','Black','Hispanic']]
                 
                 
data1980
```
